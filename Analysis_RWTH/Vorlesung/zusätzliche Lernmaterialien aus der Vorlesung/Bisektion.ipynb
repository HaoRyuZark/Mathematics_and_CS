Wir wollen ein Programm schreiben, mit dem wir eine Nullstelle mit dem Bisektionsverfahren approximieren können. Dazu benötigen wir zunächst eine Funktion, die uns das Vorzeichen von f(a)f(b) zurückgibt.

```julia
function Vorzeichen(f,a,b)
    c=(a+b)/2
    if f(c)==0
        return "Nullstelle in $c"
    end
    
    if f(c)*f(a)<0
        return (a,c)
    end
    if f(c)*f(b)<0
        return (c,b)
    end
    
end
```

```julia
(a,b)=Vorzeichen(sin(x),1,4)
```

```julia
function Nullstelle(f,a,b,epsilon)
    Schritte=1000
    if f(a)==0
        return "schon Null in a=$a"
    end
    if f(b)==0
        return "schon Null in b=$b"
    end
    if f(a)*f(b)>0
        return "you failed to bracket a zero with your choice of a and b"
    end
    
    for i=1:1000
        
        (A,B)=Vorzeichen(f,a,b)
        a=A
        b=B
        if abs(f(a))<epsilon
            return "approximative Nullstelle in $a gefunden mit $i Schritten, Fehlertoleranz $epsilon"
        end
        if abs(f(b))<epsilon
            return "approximative Nullstelle in $b gefunden mit $i Schritten, Fehlertoleranz $epsilon"
        end
    end
    return "nicht konvergiert in $Schritte Schritten"
end
```

```julia
Nullstelle((x^2-2)*sin(x),1,3,0.00001)
```

```julia
Nullstelle(sin(x),1,4,0.0000001)
```
