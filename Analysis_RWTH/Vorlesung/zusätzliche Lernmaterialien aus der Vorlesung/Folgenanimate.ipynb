### Einige Zahlenfolgen

```julia
using SymPy
using Plots
```

```julia
function c(n)
        cn=sqrt(n+1)-sqrt(n)
    return cn
end
```

```julia
function cplot(n)
    x= 1:n;
    y=[c(i) for i in 1:n]
    scatter(x,y,title="c(n)",xlim=(0,40),ylim=(0,1),xlabel="n",ylabel="c(n)",label="c(n)",color="red",markersize=4)
end
```

```julia
anim = @animate for i=1:40
cplot(i) # each frame is given by the plot
end
gif(anim, fps = 2,loop=1)
```

```julia
function a(n)
    an=1/2*(1+(-1)^n)+1/n
    #an=(-1)^n
    #an=(-1)^n/n
    return an
end
```

```julia
function aplot(n)
    x=1:n;
    y=[a(i) for i=1:n]
    scatter(x,y,title="a(n)",xlim=(0,40),ylim=(0,2),xlabel="n",ylabel="a(n)",label="a(n)",color="blue",markersize=4)
    #scatter(x,y,title="a(n)",xlim=(0,40),ylim=(-1.1,1.1),xlabel="n",ylabel="a(n)",label="a(n)",color="blue",markersize=4)
end
```

```julia
anim = @animate for i=1:40
aplot(i) # each frame is given by the plot
end
gif(anim, fps = 1)
```
